```python
# 3.1 (12 points) Let u be a Poisson number with λu = 46.2 and let v be the sum of 20 Poissonian
# numbers with λv = 2.31.
# • Generate 1000 random numbers according to u and v and plot these illustratively.
# • Test if these two distributions are statistically equal, and argue if this is reasonable or not.
# • Now generate 1000 random numbers according to w, the sum of two Binomial numbers,
# each with N = 100 and p = 0.231. To what extend is this distribution alike with the two
# previous ones?
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#3.1.1
u = np.random.poisson(lam=46.2, size=1000)
v = np.sum(np.random.poisson(lam=2.31, size=(20, 1000)), axis=0)

plt.hist(u, bins=20, alpha=0.5, label='u')
plt.hist(v, bins=20, alpha=0.5, label='v')
plt.legend(loc='upper right')
plt.show()
```

```python
#3.1.2
#The Kolmogorov-Smirnov test is used to compare the 2 distributions.
from scipy.stats import ks_2samp
stat, pvalue = ks_2samp(u, v)
print(f"Test statistic: {stat:.4f}")
print(f"P-value: {pvalue:.4f}")
```

```python
#3.1.3
w = np.sum(np.random.binomial(n=100, p=0.231, size=(2, 1000)), axis=0)

plt.hist(u, bins=20, alpha=0.5, label='u')
plt.hist(v, bins=20, alpha=0.5, label='v')
plt.hist(w, bins=20, alpha=0.5, label='w')
plt.legend(loc='upper right')
plt.show()

stat_uv, pvalue_uv = ks_2samp(u, v)
stat_uw, pvalue_uw = ks_2samp(u, w)
stat_vw, pvalue_vw = ks_2samp(v, w)

print(f"u vs. v: Test statistic = {stat_uv:.4f}, p-value = {pvalue_uv:.4f}")
print(f"u vs. w: Test statistic = {stat_uw:.4f}, p-value = {pvalue_uw:.4f}")
print(f"v vs. w: Test statistic = {stat_vw:.4f}, p-value = {pvalue_vw:.4f}")
```
