```python
# 1.2 (8 points) A store has 53.9 customers/day, and considers the top 20% busiest days to be. . . busy!
# • What distribution should the number of daily cunstomers follow?
# • Discuss what number of customers constitutes a busy day.
# • What is the average number of customers on a busy day?
```

```python
#follows a Poisson distribution, but given that lambda is 53.9, it shifts into a Gaussian distribution
#1.2.2
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import math
lam=53.9
percent=0.8
std=math.sqrt(lam)
p=norm.ppf(percent,lam, std)
print("The number of costumers that constitutes a busy day is: "+str(p)+".")
```

```python
#1.2.3

x = np.arange(20, 90, 0.5)
y = norm.pdf(x, lam, std)

plt.plot(x, norm.pdf(x,lam,std))
plt.xlabel('Costumers ')
plt.ylabel('Frequency')
plt.title('Gaussian Distribution')

plt.fill_between(x, norm.pdf(x,lam,std), where=(x>=60), color='g')
plt.show()
 
```
